Agora, carreguemos o arquivo Excel com a ajuda da biblioteca.

```python
# Importando a biblioteca Openpyxl
import openpyxl
```

Utilizarei um arquivo com dados do Airbnb na Área Metropolitana de Lisboa por realizar anteriormente [um projeto com o dataset](https://bit.ly/Analise_Airbnb_AreaMetropolitanaDeLisboa). A primeira folha no workbook contém informações gerais sobre o nome do alojamento, nome do proprietário, etc.  A outra folha é uma folha em branco para teste e amostra.

Agora, crie o objeto variável para carregar o arquivo Excel nele para outras operações.

```python
# Entre parênteses, precisamos escrever o caminho da folha de cálculo 
wb = openpyxl.load_workbook("Excel.xlsx")
```

##**Lendo informações básicas da folha de cálculo**

O wb é a variável com a qual podemos realizar várias operações. Agora, leremos várias informações sobre a folha de cálculo com o método de **propriedades**.

```python
wb.properties
```

Nesta saída, obtemos tantas informações como:

*   Criador do workbook.
*   Data de criação e data de modificação do workbook.

##**Localizando o número de folhas de cálculo e folhas de cálculo ativas no workbook**

```python
wb.sheetnames
```

A saída mostra o nome dos nomes das folhas de cálculo na nossa folha de cálculo, conforme também mostrado na imagem acima.

```python
wb.active
```

##**Criando uma folha de cálculo e removendo a folha de cálculo do python**

###**Criando folha de cálculo**

```python
# Criando folha de cálculo chamada 'new_sheet'
wb.create_sheet('new_sheet')
```

Para fazer as alterações, precisamos salvar a folha de cálculo.

```python
wb.save("Excel.xlsx")
```

New_Sheet.png

###**Removendo a nova ou qualquer folha**

Para remover a folha de cálculo, temos que dar o nome dessa folha de cálculo em particular.

Opção 1

```python
folha = wb['new_sheet'] 
wb.remove(folha)
```

Opção 2

```python
del wb['new_sheet']
```

Podemos usar qualquer método das opções acima.

##**Métodos de trabalho com folhas de cálculo**

Escolha o nome da folha de cálculo em que queremos trabalhar.

```python
sheet = wb['listings'] 
sheet.title
```

Conhecer todos os métodos para efetuar operações com folhas de cálculo com ***o método dir***. O método ***dir*** mostra todos os métodos com os quais podemos elaborar manipulações na folha de cálculo.

```python
dir(sheet)
```

Exemplos:

```python
sheet.active_cell
```

```python
sheet.dimensions
```

A célula ativa significa que a seta do rato clicou na célula D14. A área total preenchida da folha é de A1 a R19715.

```python
# Número da última linha com valores
sheet.max_row
```

```python
# Número da última coluna com valores
sheet.max_column
```

```python
# Ver toda a folha de cálculo
for i in sheet.values:
    print(i)
```

##**Encontrando os valores na folha de cálculo**

Podemos encontrar os dados da célula pelo nome da célula ou mencionando os valores das linhas e colunas.

```python
# Através do nome da célula
sheet['A4'].value
```

```python
# Através do valor la linha e coluna
sheet.cell(row = 3, column = 5).value
```

##**Alterando o valor da célula**

Também podemos alterar o valor dos dados da célula conforme o exemplo abaixo:

```python
cell = sheet['D4']
cell.value = 'Arthur Henke'
wb.save("Excel.xlsx")
```

Célula Alterada.png

##**Estilização das células**

Estamos a trabalhar em Excel e esquecemos do estilo das células e dos dados. Aqui, podemos elaborar muitos estilos como negrito, itálico, cor da fonte, cor de preenchimento da célula, etc.

```python
from openpyxl.styles import Font
cell = sheet['A5']
font = Font(color='FFFF0000', bold = True, italic = True)
cell.font = font
wb.save("Excel.xlsx")
```
